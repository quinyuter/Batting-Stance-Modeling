```python
## Libraries
import pandas as pd
import numpy as np
import matplotlib as ply
import seaborn as sns
```

```python
## Downloading Data
stance = pd.read_csv("/Users/quinyuter/Desktop/Bat Stance Project/batting-stance.csv")
ev = pd.read_csv("/Users/quinyuter/Desktop/Bat Stance Project/exit_velocity (1).csv")
bat_stats = pd.read_csv("/Users/quinyuter/Desktop/Bat Stance Project/stats (1).csv")
```

```python
## Merging Data
# Clean column names
ev.columns = ev.columns.str.strip()
bat_stats.columns = bat_stats.columns.str.strip()

# Rename player name column for consistency (optional)
ev.rename(columns={"last_name, first_name": "name"}, inplace=True)
bat_stats.rename(columns={"last_name, first_name": "name"}, inplace=True)

# Merge all on player_id
merged_df = stance.merge(ev, left_on='id', right_on='player_id', how='left')
merged_df = merged_df.merge(bat_stats, on='player_id', how='left')
```

```python
# Clean column names
def clean_column_names(df):
    df.columns = (
        df.columns.str.strip()
                  .str.lower()
                  .str.replace('[^0-9a-zA-Z]+', '_', regex=True)
                  .str.strip('_')
    )
    return df

data = clean_column_names(merged_df)

# Rename b_rbi to rbi
data.rename(columns={'b_rbi': 'rbi', 'name_x': 'name'}, inplace=True)

# Drop only unnecessary columns (keep 'name' and 'rbi')
columns_to_drop = ['name_y', 'unnamed_0']
data.drop(columns=columns_to_drop, errors='ignore', inplace=True)
data.head()
```
